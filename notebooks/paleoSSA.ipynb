## Data: the "updated benthic stack"

Accesed through https://github.com/seonminahn/HMM-Stack

[1] A probabilistic Pliocene-Pleistocene stack of benthic δ18O using a profile hidden Markov model

```python
### get data
stack = pd.read_csv('https://raw.githubusercontent.com/seonminahn/HMM-Stack/master/Prob_stack.txt',sep='  ',engine='python',header=None)
stack.rename(columns={0:'age',1:'d18o',2:'std',3:'c95',4:'c5'},inplace=True)
stack = stack.set_index('age')
 
```

```python
### linear timing?
dt_or = np.diff(stack.index.values) # time difference 
f,ax = plt.subplots(1,1)
ax.plot(stack.index[:-1],dt_or,color='xkcd:dark grey')
ax.set_xlim(0,5000)
ax.set_ylabel('$\mathrm{\delta^{18}O_{benthic}\ (^o/_{oo})}$')
ax.set_xlabel('Time (kyr BP)')

dt_or.mean()
```

```python
from scipy.interpolate import interp1d

t_new = np.arange(0,3000,2.5)
stack_lin = pd.DataFrame(index=t_new)
stack_lin['d18o'] = interp1d(stack.index.values,stack['d18o'])(t_new)
stack_lin['std'] = interp1d(stack.index.values,stack['std'])(t_new)

stack_lin.plot()
```

## Data: the Antarctic composite CO2 record

Accesed through https://www.ncei.noaa.gov/

[1] Revision of the EPICA Dome C CO2 record from 800 to 600 kyr before present(https://doi.org/10.1002/2014GL061957)

```python
url = 'https://www.ncei.noaa.gov/pub/data/paleo/icecore/antarctica/antarctica2015co2composite.txt'
df = pd.read_csv(url,header=137,sep='\t')
df['age_gas_calBP'] = df['age_gas_calBP']/1000
t_new = np.arange(0,800,0.5)
df_ = pd.DataFrame(index=t_new)
df_['co2'] = interp1d(df['age_gas_calBP'],df['co2_ppm'])(t_new)
df_['std'] = interp1d(df['age_gas_calBP'],df['co2_1s_ppm'])(t_new)

df_.plot()
```

```python
#### SSA
ssa_d18o = util.SSA(stack_lin['d18o'], L=200)

f,ax = plt.subplots(1,1)
ax.plot(ssa_d18o.reconstruct([0]))
ax.plot(ssa_d18o.reconstruct([0,1,2,3,4]))
```

```python
from scipy import signal 
def next_power_of_2(x):  
    return 1 if x == 0 else 2**(x - 1).bit_length()

def periodogram(x,y):
    fs = 1/(np.mean(np.diff(x)))
    f,Pxx = signal.periodogram(y,fs=fs,nfft=next_power_of_2(len(y)))
    return f,Pxx


f,pxx = periodogram(ssa_d18o.reconstruct([2]).index,ssa_d18o.reconstruct([3]))
plt.plot(f,pxx)
plt.xlim(0,0.05)
```
